# AGE GROUPS

```python
basegb.dtypes
```

```python
#subset all the ages and call it age_years 

age_yrs = [basegb['age_yrs']]
age_yrs
```

```python
#makes the dataframe a float
basegb["age_yrs"] = pd.to_numeric(basegb["age_yrs"], downcast="float") 
```

```python
# create a user defined function for age groups
def age_grp(basegb):
   if (basegb.age_yrs >=25 and basegb.age_yrs<40): return 1
   elif (basegb.age_yrs >=40 and basegb.age_yrs<65): return 2
   elif (basegb.age_yrs>=65): return 3
   else: return 0 
```

```python
#apply the user defined function age_grp on the column name age_yrs and create a new column called agegrps
basegb['agegrps']=basegb.apply(age_grp, axis=1)
```

```python
#to test the effectiveness of your definition, run the range of the groups
mins=basegb.groupby('agegrps').age_yrs.min()
mins
means=basegb.groupby('agegrps').age_yrs.mean()
maxs=basegb.groupby('agegrps').age_yrs.max()

allranges=pd.concat([mins,means,maxs], axis=1)
allranges
```
